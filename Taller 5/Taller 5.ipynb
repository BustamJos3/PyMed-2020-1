### 1.  Dada una lista, obtener el mínimo y el máximo usando numpy.

```python
lista = [1,2,3,4,10,20]
array = np.array(lista)

print('El mínimo es {}'.format(array.min()))
print('El máximo es {}'.format(array.max()))
```

### 3.Crear un diccionario de dos niveles donde el nivel uno toma las llaves ['A', 'B'] que representan dos tipos de combustibles, y para cada combustible se tiene un diccionario con 7 llaves [1,2,3,4,5,6,7] que corresponde a una prueba. Cada una de estas pruebas tiene asocidado un DataFrame vacío.

```python
diccionario = {
               'A':{
                   1:pd.DataFrame(),
                   2:pd.DataFrame(),
                   3:pd.DataFrame(),
                   4:pd.DataFrame(),
                   5:pd.DataFrame(),
                   6:pd.DataFrame(),
                   7:pd.DataFrame(),
                   },
                'B':{
                    1:pd.DataFrame(),
                   2:pd.DataFrame(),
                   3:pd.DataFrame(),
                   4:pd.DataFrame(),
                   5:pd.DataFrame(),
                   6:pd.DataFrame(),
                   7:pd.DataFrame(),
                    }
                }

diccionario
```

```python
diccionario = {}
for fuel in ['A','B']:
    diccionario[fuel] = {}
    for test in range(1,8):
        diccionario[fuel][test] = pd.DataFrame()
diccionario
```
